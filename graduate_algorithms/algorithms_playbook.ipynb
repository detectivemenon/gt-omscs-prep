## Dynamic Programming — Example: Fibonacci

```python
def fib_dp(n):
    if n <= 1:
        return n
    a, b = 0, 1
    for _ in range(2, n+1):
        a, b = b, a+b
    return b

# TODO: add memoized recursive version and compare times
```
